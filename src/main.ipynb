```python
import numpy as np
def create_X_y(scaled_mag, scaled_delta_t, window_len):
    X, y = [], []
    for i in range(0, np.shape(scaled_mag)[0] - window_len):
        features = []
        for j in range(i, i + window_len):
            feature = np.concatenate((scaled_mag[j], scaled_delta_t[j]), axis=0)
            features.append(feature)
        X.append(features)
        y.append(scaled_mag[i + window_len])

    X = np.array(X)
    y = np.array(y)

    return X, y
```

```python
window_len = 3
scaled_mag = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
t = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
scaled_delta_t = np.array([1, 1, 1, 1, 1]).reshape(-1, 1)
t_pred = t[window_len:]
X = create_X_y()
```

```python
def recursive(self, t_pred, X, mag_scaler):
    delta_t_pred = np.diff(t_pred)
    scaled_y_pred = []
    X_step = X[[0]]  # shape = (1, window_len, 2)
    scaled_y_pred.append(X[0, -1, 0])
    for i in range(len(delta_t_pred)):
        scaled_y_step = self.model.predict(X_step)  # shape = (1, 1)
        scaled_y_pred.append(scaled_y_step[0, 0])  # scaled_y_step[0] -> shape = (num_feature)
        if i == len(delta_t_pred) - 1:
            break

        scaled_t_step = np.expand_dims(delta_t_pred[i], axis=0)  # shape = (1, 1)
        new_feature = np.concatenate((scaled_y_step, scaled_t_step), axis=1)  # shape = (1, 2)
        X_step = np.concatenate((X_step[0][1:, :], new_feature), axis=0)  # shape = (window_len, 2)
        X_step = np.expand_dims(X_step, axis=0)  # shape = (1, window_len, 2)

    scaled_y_pred = np.array(scaled_y_pred)  # shape = (num_step, 1)
    y_pred = mag_scaler.inverse_transform(scaled_y_pred)

    return y_pred
```
